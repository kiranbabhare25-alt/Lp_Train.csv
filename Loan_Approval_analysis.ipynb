```python
import pandas as pd 
import seaborn as sb
import matplotlib.pylab as plt
import numpy as np
```

```python
df=pd.read_csv("../Datasets/LP_Train.csv")
```

```python
df
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.Dependents.unique()
```

```python
df.Dependents=df.Dependents.fillna(0)
```

```python
df.Dependents=df.Dependents.replace('[+]','',regex=True).astype('int64')
```

```python
df.Gender.unique()
```

```python
df.Gender.value_counts()
```

```python
df.Gender=df.Gender.fillna('Male')
```

```python
df.Married.unique()
```

```python
df.Married.value_counts()
```

```python
df.Married=df.Married.fillna('Yes')
```

```python
df.Self_Employed.unique()
```

```python
df.Self_Employed.value_counts()
```

```python
df.Self_Employed=df.Self_Employed.fillna('Yes')
```

```python
df.LoanAmount.unique()
```

```python
df.LoanAmount.mean()
```

```python
df.LoanAmount=df.LoanAmount.fillna(146).astype('int64')
```

```python
df.Loan_Amount_Term.unique()
```

```python
df.Loan_Amount_Term.mean()
```

```python
df.Loan_Amount_Term=df.Loan_Amount_Term.fillna(342.0)
```

```python
df.Credit_History.unique()
```

```python
df.Credit_History.mean()
```

```python
df.Credit_History=df.Credit_History.fillna(0.).astype('int64')
```

```python
df=df.rename(columns={'Loan_Status':'Loan_approval'})
```

 Demographic Analysi

```python
cols=['Gender','Married','Dependents','Education','Self_Employed']

for col in cols:
    pd.crosstab(df[col],df['Loan_approval']).plot(kind='bar')
    plt.show()
```
